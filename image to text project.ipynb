```python
import pytesseract
#path of library
pytesseract.pytesseract.tesseract_cmd=r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'

#import libraries 
import pyttsx3
from PIL import Image
import numpy as np
import speech_recognition as sr
import cv2

def process_audio():
    #intro assistant
    process_audio.engine= pyttsx3.init('sapi5')
    voices= process_audio.engine.getProperty('voices')
    print(voices[1].id)
    process_audio.engine.setProperty('voice',voices[1].id)


def speak(audio):
    process_audio.engine.say(audio)
    process_audio.engine.runAndWait()

def reshape_image():
    img=cv2.imread(main.image_name)
    if img is None:
        print("image couldn't read" )
        speak("image couldn't read" )
    else:
        print("image trace")
        speak("loading image")
    print('original shape',img.shape)
    
    height= img.shape[1]
    width= img.shape[0]

    if (height!=1000):
        height=1000
    if(width!=650):
        width=650
    
    dim=(height,width)
    reshape_image.resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    cv2.imshow("Original image",reshape_image.resized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print('resized shape',reshape_image.resized.shape)
    
def text_detect():
    img=cv2.cvtColor(reshape_image.resized,cv2.COLOR_BGR2RGB)
    
    imgH,imgW,_=img.shape
    boxes= pytesseract.image_to_boxes(img)
    for b in boxes.splitlines():
        b=b.split(' ')
        print(b)
        x,y,w,h= int(b[1]),int (b[2]), int (b[3]),int (b[4]) 
        bt=cv2.rectangle(img,(x,imgH-y),(w,imgH-h),(0,0,255),1)
        np.array(b)
    ret, text_detect.img_binary = cv2.threshold(bt,100, 255, cv2.THRESH_BINARY_INV)
    speak("text detection is complete")
    cv2.imshow('detected Text',bt)
    cv2.imshow('binary_inverse',text_detect.img_binary)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def webcam():
    cam_video= cv2.VideoCapture(0)
    upper_left = (155, 0)
    bottom_right = (1120, 700)
    cam_video.set(cv2.CAP_PROP_FRAME_WIDTH, 1000)
    cam_video.set(cv2.CAP_PROP_FRAME_HEIGHT, 1000)
    print("press 's' for capture the image:")
    speak("press 's' the for capture the image:")
    print("press 'q' for close the video window:")
    speak("press 'q' for close the video window:")
    
    while(True):
        ret,frame=cam_video.read()
        r = cv2.rectangle(frame, upper_left, bottom_right, (100, 50, 200), 5)
        rect_img = r[upper_left[1] : bottom_right[1], upper_left[0] : bottom_right[0]]
        cv2.imshow('frame',rect_img)
        key= cv2.waitKey(1)
        
        
        #filename="webcamimage_1.jpg"
        #keyborad_op={"s":"save"}
        if key==ord('s'):
            cv2.imwrite(filename="webcamimage_1.jpg",img=rect_img)
            print("image is captured")
            speak("image is captured")
            cv2.destroyAllWindows()
            continue
            
        elif key==ord('q'):
            cv2.destroyAllWindows()
            break

#print the recognized text
def text_recognition():
    ret, img_binary = cv2.threshold(reshape_image.resized,100, 255, cv2.THRESH_BINARY_INV)
    text_recognition.text= pytesseract.image_to_string(img_binary)
    print(text_recognition.text)
    speak(text_recognition.text)

def save_into_file():
    with open("detected_text_by_passing.txt",'w') as file:
        file.write(text_recognition.text)
        file.close()
        

def web_cam_text_file():
    with open("webcam detected_text.txt",'w') as file:
        file.write(text_recognition.text)
        file.close()
        
def choice(x):
    if x =='1':
        print("choice 1")
        main.image_name=input('enter the name of image with extenion:' )
        reshape_image()
        print("image shaped done! ")
        text_detect()
        print("Text detection show in image.. ")
        text_recognition()
        print("detected text show in result.")
        speak("detected text show in result.")
        save_into_file()
        speak("detected text store in file")
        
    elif x=='2':
        print("choice 2")
        webcam()
        main.image_name= 'webcamimage_1.jpg'
        reshape_image()
        print("image shaped done! ")
        text_detect()
        print("Text detection show in image.. ")
        text_recognition()
        print("detected text show in result.")
        speak("detected text show in result.")
        web_cam_text_file()
        speak("detected text store in file.")
        print("detected text store in file.")
    
    elif x=='q':
        print('quit')
    else:
        print("Enter correct choice")
        speak("Enter Correct choice")
    

def main():
                
    print('press 1 if you want to detect the image_text by passing.\n'+'press 2 if you want to detect the image_text by webcam.\n'+'press q to quit the program.\n')
    x=input('enter your choice')
    choice(x)
    
if __name__=="__main__":
    process_audio()
    #speak()
    speak("Welcome to Image to Text detection Program")
    speak("enter your choice")
    
    main()
```
